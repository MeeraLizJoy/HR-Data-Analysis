## Correlation Matrix

Correlation matrix for all numeric values

```python
numeric_df = df.select_dtypes(include=['number'])

corr_matrix = numeric_df.corr()
plt.figure(figsize = (20,10))
sns.heatmap(corr_matrix, annot = True, cmap = "coolwarm")
plt.title("Correlation Map for Numeric Attributes")
plt.show()
```

Two variables are said to be highly correlation when they have a value of 0.7 or greater. 

The correlation matrix shows that the attributes "MonthlyIncome" and "Joblevel" are highly correlated - 0.95 correlation.

Also, "JobLevel" and "Total WorkingYears" - 0.78

"PercentSalaryHike" and RelationshipSatisfaction" - 0.77

```python
fig, ax = plt.subplots(2, 2, figsize=(10, 10))
plt.suptitle("Understanding the distribution of various factors", fontsize=20)
sns.histplot(data=df, x='Age', ax=ax[0, 0])
ax[0, 0].set_title('Distribution of Age')
sns.histplot(data=df, x='TotalWorkingYears', ax=ax[0, 1])
ax[0, 1].set_title('Distribution of Total Working Years')
sns.histplot(data=df, x='YearsAtCompany', ax=ax[1, 0])
ax[1, 0].set_title('Distribution of Years at company')
sns.histplot(data=df, x='YearsInCurrentRole', ax=ax[1, 1])
ax[1, 1].set_title('Distribution of Years in Current Role')
plt.show()
```

Majority of the employees are in the age group of 35-40 years.
They have a total experience of around 8-10 years. 
They have typically spent around 2-4 years at the current organization and have spent 2-3 years in the current role they are in.

## Data Visualizations

### 1. Attrition

```python
df['Attrition_numeric'] = df['Attrition'].map({'Yes': 0, 'No': 1})
sns.countplot(data=df, x='Attrition_numeric')
plt.show()
```

The plot shows that around 250 people have left the organization while the organization has been able to retain around 1200 people.

### 2.Overtime

```python
sns.countplot(data = df, x = "OverTime")
plt.title("Over Time")
plt.show()
```

### 3. Marital Status

```python
from numpy import median
sns.barplot(x='Attrition', y='MonthlyIncome', hue= 'MaritalStatus',data=df, estimator=median)
plt.title("Marital Status")
plt.show()
```

### 4. Job Role

```python
sns.countplot(data = df, x = "JobRole", hue = "Attrition")
plt.title("Job Role")
plt.xticks(rotation=90)
plt.show()
```

### 5. Gender

```python
sns.countplot(data = df, x = "Gender")
plt.title("Gender")
plt.show()
```

### 6. Education Field

```python
education_counts = df['EducationField'].value_counts()


plt.figure(figsize=(10, 8))
plt.pie(education_counts, labels=education_counts.index, autopct='%1.1f%%', startangle=140)
plt.axis('equal')  
plt.title("Education Field Distribution")
plt.show()
```

### 6. Department

```python
import squarify
department_counts = df['Department'].value_counts()

# Create a squarify treemap
plt.figure(figsize=(12, 8))
squarify.plot(sizes=department_counts.values, label=department_counts.index, alpha=0.6)
plt.axis('off')
plt.title("Department-wise Employee Distribution (Treemap)")
plt.show()
```

### 7. Buisness Travel

```python
travel_counts = df['BusinessTravel'].value_counts()

# Create a donut chart
plt.figure(figsize=(8, 8))
plt.pie(travel_counts, labels=travel_counts.index, autopct='%1.1f%%', startangle=90, wedgeprops=dict(width=0.3))
plt.axis('equal')  # Equal aspect ratio ensures that pie is drawn as a circle
plt.title("Business Travel Frequency")
# Add a circle at the center to make it a donut chart
centre_circle = plt.Circle((0,0),0.70,fc='white')
fig = plt.gcf()
fig.gca().add_artist(centre_circle)
plt.show()
```

### 10. Relation between Overtime and Age

```python
f,ax = plt.subplots(figsize=(15,10))
sns.boxplot(x='JobSatisfaction', y='MonthlyIncome', data=df, hue='JobLevel',palette='Set3')
plt.legend(loc='best')
plt.show()
```

From this boxplot we can observe the difference in monthly income across different job levels. The difference in the monthly income is observed by the median value for different job satisfaction level.
